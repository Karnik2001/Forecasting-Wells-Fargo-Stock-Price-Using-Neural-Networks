```python
# Import all of the necessary libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn import metrics
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import *
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from keras.callbacks import EarlyStopping
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
# Get the current date and time
end = datetime.now()
# Calculate the start date (5 years ago)
start = datetime(end.year - 10, end.month, end.day)
# Download historical stock data for Wells Fargo ('WFC')
wells = yf.download('WFC', start, end)
```

```python
# Extract the month from the Date index and create a new 'month' column
wells['month'] = wells.index.month

# Now you can use the 'month' column to create the 'is_quarter_end' column
wells['quarter'] = np.where(wells['month']%3==0,1,0)

# Observe the dataset
wells.head()
```

```python
# Create new features:
wells['open-close']  = wells['Open'] - wells['Close']  # Difference between opening and closing prices
wells['low-high']  = wells['Low'] - wells['High']  # Difference between low and high prices

# Create the target variable for prediction:
wells['target'] = np.where(wells['Close'].shift(-1) > wells['Close'], 1, 0)

# If tomorrow's closing price is higher than today's, target is 1, otherwise 0
```

```python
# Define features and target variable
features = wells[['open-close', 'low-high', 'quarter']]  # Select the desired columns as features
target = wells['target']  # Select the 'target' column as the target variable

# Handle missing values (NaNs) using forward fill
wells = wells.fillna(method='ffill')  # Fill NaNs with the previous valid value

# Scale features using StandardScaler
scaler = StandardScaler()  # Create a StandardScaler object
features = scaler.fit_transform(features)  # Fit the scaler to the features and transform them

# Split data into training and validation sets
X_train, X_valid, Y_train, Y_valid = train_test_split(
    features, target, test_size=0.2, random_state=2022)
    # Split the data with 20% for validation and a fixed random state for reproducibility

# Print the shapes of the training and validation sets
print(X_train.shape, X_valid.shape)  # Display the dimensions of the resulting datasets
```

```python
# Define a list of models to evaluate
models = [
    LogisticRegression(),  # Logistic Regression model
    SVC(kernel='poly', probability=True),  # Support Vector Machine with a polynomial kernel
    XGBClassifier()  # XGBoost Classifier
]

# Iterate through each model in the list
for i in range(3):
    # Train the current model on the training data
    models[i].fit(X_train, Y_train)

    # Print the model name
    print(f'{models[i]} : ')

    # Calculate and print the training accuracy using ROC AUC score
    print('Training Accuracy : ', metrics.roc_auc_score(
        Y_train, models[i].predict_proba(X_train)[:, 1]))

    # Calculate and print the validation accuracy using ROC AUC score
    print('Validation Accuracy : ', metrics.roc_auc_score(
        Y_valid, models[i].predict_proba(X_valid)[:, 1]))

    # Print a blank line to separate the output for each model
    print()
```

```python
model = Sequential()

#Adding the first LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
model.add(Dropout(0.1))

# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.1))

# Adding a third LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.1))

# Adding a fourth LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50))
model.add(Dropout(0.1))

# Adding the output layer with sigmoid activation
model.add(Dense(units = 1, activation='relu'))

# Compiling the RNN with binary_crossentropy loss
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# Predict the value
y_pred = model.predict(X_valid)
```

```python
y_pred = model.predict(X_valid)

# Convert probabilities to binary predictions using a threshold (e.g., 0.5)
y_pred_binary = (y_pred > 0.5).astype(int)

# Now you can calculate the accuracy
y_pred = model.predict(X_valid)

# Convert probabilities to binary predictions using a threshold (e.g., 0.5)
y_pred_binary = (y_pred > 0.5).astype(int)

# Now you can calculate the accuracy
accuracy_score(Y_valid, y_pred_binary)
r2_score(Y_valid,y_pred_binary)

# Print the scores
print(f'R2 score: {r2_score(Y_valid,y_pred_binary)}')
print(f'Accuracy score : {accuracy_score(Y_valid, y_pred_binary)}')
print(f'Mean Absolute Error : {mean_absolute_error(Y_valid, y_pred_binary)}')
print(f'Mean Squared Error : {mean_squared_error(Y_valid, y_pred_binary)}')
```
